```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris=load_iris()
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head())
```

```python
iris.target_names
```

Add a target colum

```python
df['target']=iris.target
df.head()
```

if we want to change the value use this


```python
df[df.target==1].head() #we use 1 ,2,3 what we want
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
df.head()
```

slicing the value ...ass the value in 0  to 45 : 55

```python
df[45:55]
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

sepal length vs sepal width(setosa vs versicolor

```python
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='.')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='red',marker='*')
```

```python
plt.xlabel('petal width')
plt.ylabel('petal length')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='green',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='.')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='red',marker='*')
```

```python
from sklearn.model_selection import train_test_split
```

  if you need to drop
  

```python
x=df.drop(['target','flower_name'],axis='columns')
y=df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train,y_train)
```

Regularization

```python
model_C=SVC(C=1)
model_C.fit(x_train,y_train)
model_C.score(x_test,y_test)
```

```python
model = SVC(C=10)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

Gamma

```python
model_g=SVC(gamma=10)
model_g.fit(x_train,y_train)
model_g.score(x_test,y_test)
```

Kernel

```python
model_kernel=SVC(kernel='linear')
model_kernel.fit(x_train,y_train)
#model_kernel.score(x_test,y_test)
```

```python
model_kernel.score(x_test,y_test)
```
